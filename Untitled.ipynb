```python
# ============================================================
# PIPELINE COMPLETO TFM (GSE2034 ER+ + VALIDACIÓN EXTERNA)
# ============================================================

import numpy as np
import pandas as pd
import GEOparse
import matplotlib.pyplot as plt
import time

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import VarianceThreshold, SelectKBest, f_classif, RFE
from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import mutual_info_classif
from sklearn.metrics import roc_auc_score, roc_curve, confusion_matrix

print("\n=== CONFIGURACIÓN ===")

RANDOM_STATE = 42
K_VALUES = [20, 50, 100, 500, 1000]


# ============================================================
# A: CARGA GSE2034 (EXPRESIÓN)
# ============================================================

print("\n=== CARGANDO GSE2034 ===")

gse = GEOparse.get_GEO("GSE2034", destdir="../data")

expr_data = {}
for gsm_name, gsm in gse.gsms.items():
    expr_data[gsm_name] = gsm.table.set_index("ID_REF")["VALUE"]

X = pd.DataFrame(expr_data).T
print("Shape expresión:", X.shape)


# ============================================================
# A2: CARGA CLÍNICA (TU CSV)
# ============================================================

print("\n=== CARGANDO CLÍNICA ===")

clinical = pd.read_csv(
    "data/GSE2034_clinical.csv",
    sep="\t",
    comment="#",
    engine="python"
)

# Asegurar nombres
clinical = clinical.rename(columns={
    "GEO asscession number": "GSM",
    "relapse (1=True)": "relapse"
})

print("Shape clínica:", clinical.shape)


# ============================================================
# A3: MERGE EXPRESIÓN + CLÍNICA
# ============================================================

print("\n=== MERGE ===")

X["GSM"] = X.index

df = X.merge(clinical, on="GSM")

print("Shape merged:", df.shape)


# ============================================================
# A4: FILTRADO ER+
# ============================================================

print("\n=== FILTRANDO ER+ ===")

df = df[df["ER Status"] == "ER+"].copy()

print("Shape ER+:", df.shape)


# ============================================================
# A5: MATRICES FINALES
# ============================================================

y = df["relapse"].astype(int).values

X = df.drop(columns=[
    "GSM", "PID", "relapse",
    "ER Status", "lymph node status",
    "time to relapse or last follow-up (months)",
    "Brain relapses (1=yes, 0=no)"
])

print("Shape X final:", X.shape)
print("Distribución:", np.bincount(y))


# ============================================================
# A6: LOG TRANSFORM
# ============================================================

X_log = np.log2(X + 1)


# ============================================================
# B: EXPERIMENTACIÓN
# ============================================================

print("\n=== EXPERIMENTO K ===")

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

for K in K_VALUES:

    print(f"\nK = {K}")

    pipelines = {

        "LR baseline": Pipeline([
            ("variance", VarianceThreshold(0.01)),
            ("scaler", StandardScaler()),
            ("model", LogisticRegression(max_iter=1000))
        ]),

        "LR + ANOVA": Pipeline([
            ("variance", VarianceThreshold(0.01)),
            ("scaler", StandardScaler()),
            ("select", SelectKBest(f_classif, k=K)),
            ("model", LogisticRegression(max_iter=1000))
        ]),

        "LR + MI": Pipeline([
            ("variance", VarianceThreshold(0.01)),
            ("scaler", StandardScaler()),
            ("select", SelectKBest(mutual_info_classif, k=K)),
            ("model", LogisticRegression(max_iter=1000))
        ]),

        "LR + RFE": Pipeline([
            ("variance", VarianceThreshold(0.01)),
            ("scaler", StandardScaler()),
            ("select", RFE(LogisticRegression(max_iter=1000), n_features_to_select=K, step=50)),
            ("model", LogisticRegression(max_iter=1000))
        ])
    }

    for name, pipe in pipelines.items():

        scores = cross_validate(
            pipe,
            X_log,
            y,
            cv=cv,
            scoring="roc_auc",
            n_jobs=2
        )

        print(f"{name} → AUC: {np.mean(scores['test_score']):.4f}")


# ============================================================
# C: MODELO FINAL (L1)
# ============================================================

print("\n=== MODELO FINAL ===")

l1_pipe = Pipeline([
    ("variance", VarianceThreshold(0.01)),
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(
        penalty="l1",
        solver="saga",
        C=0.1,
        max_iter=5000
    ))
])

l1_pipe.fit(X_log, y)

coefs = l1_pipe.named_steps["model"].coef_[0]
mask = l1_pipe.named_steps["variance"].get_support()
genes_after_var = X_log.columns[mask]

selected_genes = genes_after_var[coefs != 0]

print("Probes seleccionados:", len(selected_genes))


# ============================================================
# C2: MAPPING
# ============================================================

platform = gse.gpls[list(gse.gpls.keys())[0]]
annot = platform.table

map_dict = dict(zip(annot['ID'], annot['Gene Symbol']))

mapped_selected_genes = [
    map_dict.get(g) for g in selected_genes
]

mapped_selected_genes = list(dict.fromkeys([
    g for g in mapped_selected_genes if g is not None
]))

print("Genes finales:", mapped_selected_genes)


# ============================================================
# D: VALIDACIÓN EXTERNA (GSE58984)
# ============================================================

print("\n=== VALIDACIÓN EXTERNA ===")

gse_ext = GEOparse.get_GEO("GSE58984", destdir="../data")

# EXPRESIÓN
expr_data = {}
for gsm_name, gsm in gse_ext.gsms.items():
    expr_data[gsm_name] = gsm.table.set_index("ID_REF")["VALUE"]

X_ext = pd.DataFrame(expr_data).T
X_ext_log = np.log2(X_ext + 1)

# LABELS
y_ext = []
for gsm_name in X_ext.index:
    gsm = gse_ext.gsms[gsm_name]
    chars = gsm.metadata.get("characteristics_ch1", [])

    label = None
    for c in chars:
        if "distant metastasis" in c.lower():
            label = int(c.split(":")[-1].strip())

    if label is not None:
        y_ext.append(label)

y_ext = np.array(y_ext)
print("Distribución externa:", np.bincount(y_ext))


# ============================================================
# MAPPING EXTERNO
# ============================================================

platform_ext = gse_ext.gpls[list(gse_ext.gpls.keys())[0]]
annot_ext = platform_ext.table

map_ext = dict(zip(annot_ext['ID'], annot_ext['Gene Symbol']))

X_ext = X_ext_log.rename(columns=map_ext)

X_ext.columns = [
    str(g).split("///")[0].strip() if g is not None else None
    for g in X_ext.columns
]

X_ext = X_ext.loc[:, X_ext.columns.notna()]
X_ext = X_ext.T.groupby(level=0).mean().T


# ============================================================
# TRAIN PROCESADO
# ============================================================

mapped_clean = [
    g.split("///")[0].strip() if isinstance(g, str) else g
    for g in mapped_selected_genes
]

X_train = X_log[selected_genes].copy()
probe_to_gene = dict(zip(selected_genes, mapped_clean))
X_train = X_train.rename(columns=probe_to_gene)
X_train = X_train.T.groupby(level=0).mean().T


# ============================================================
# INTERSECCIÓN
# ============================================================

common_genes = [g for g in mapped_clean if g in X_ext.columns]

print("Genes comunes:", len(common_genes))

X_train = X_train[common_genes]
X_test = X_ext.reindex(columns=common_genes)


# ============================================================
# MODELO FINAL
# ============================================================

model = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(max_iter=1000, class_weight="balanced"))
])

model.fit(X_train, y)

y_pred = model.predict_proba(X_test)[:, 1]

auc = roc_auc_score(y_ext, y_pred)
auc_inv = roc_auc_score(y_ext, 1 - y_pred)

if auc_inv > auc:
    y_pred = 1 - y_pred
    auc = auc_inv

print("\n=== RESULTADOS ===")
print("AUC:", auc)

tn, fp, fn, tp = confusion_matrix(y_ext, (y_pred > 0.5)).ravel()

print("Sens:", tp/(tp+fn))
print("Spec:", tn/(tn+fp))


# ============================================================
# ROC
# ============================================================

fpr, tpr, _ = roc_curve(y_ext, y_pred)

plt.plot(fpr, tpr, label=f"AUC={auc:.2f}")
plt.plot([0,1],[0,1], linestyle=":")
plt.legend()
plt.grid()
plt.show()
```

```python
# ============================================================
# B: EXPERIMENTACIÓN COMPLETA (COMPARACIÓN JUSTA DE MODELOS)
# ============================================================

from sklearn.model_selection import cross_val_predict

print("\n=== EXPERIMENTO K ===")

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

for K in K_VALUES:

    print(f"\n==============================")
    print(f"K = {K}")
    print(f"==============================")

    # ====================================================
    # MODELOS DEPENDIENTES DE K
    # ====================================================

    pipelines = {

        # ------------------------------------------------
        # Modelo baseline (sin selección de características)
        # ------------------------------------------------
        "LR baseline": Pipeline([
            ("variance", VarianceThreshold(0.01)),
            ("scaler", StandardScaler()),
            ("model", LogisticRegression(
                max_iter=1000,
                class_weight="balanced"
            ))
        ]),

        # ------------------------------------------------
        # Selección ANOVA
        # ------------------------------------------------
        "LR + ANOVA": Pipeline([
            ("variance", VarianceThreshold(0.01)),
            ("scaler", StandardScaler()),
            ("select", SelectKBest(f_classif, k=K)),
            ("model", LogisticRegression(
                max_iter=1000,
                class_weight="balanced"
            ))
        ]),

        # ------------------------------------------------
        # Selección Mutual Information
        # ------------------------------------------------
        "LR + MI": Pipeline([
            ("variance", VarianceThreshold(0.01)),
            ("scaler", StandardScaler()),
            ("select", SelectKBest(mutual_info_classif, k=K)),
            ("model", LogisticRegression(
                max_iter=1000,
                class_weight="balanced"
            ))
        ]),

        # ------------------------------------------------
        # Selección RFE
        # ------------------------------------------------
        "LR + RFE": Pipeline([
            ("variance", VarianceThreshold(0.01)),
            ("scaler", StandardScaler()),
            ("select", RFE(
                LogisticRegression(max_iter=1000, class_weight="balanced"),
                n_features_to_select=K,
                step=50
            )),
            ("model", LogisticRegression(
                max_iter=1000,
                class_weight="balanced"
            ))
        ])
    }

    # ====================================================
    # EVALUACIÓN DE MODELOS CON K
    # ====================================================

    for name, pipe in pipelines.items():

        print(f"\n--- {name} ---")

        # ----------------------------
        # 1. CROSS-VALIDATION
        # ----------------------------
        start = time.time()

        scores = cross_validate(
            pipe,
            X_log,
            y,
            cv=cv,
            scoring={
                "roc_auc": "roc_auc",
                "f1": "f1",
                "accuracy": "accuracy"
            },
            n_jobs=2
        )

        end = time.time()

        print(
            f"AUC: {np.mean(scores['test_roc_auc']):.4f} | "
            f"F1: {np.mean(scores['test_f1']):.4f} | "
            f"ACC: {np.mean(scores['test_accuracy']):.4f} | "
            f"Tiempo: {end-start:.2f}s"
        )

        # ----------------------------
        # 2. MATRIZ DE CONFUSIÓN
        # ----------------------------
        y_pred = cross_val_predict(pipe, X_log, y, cv=cv)

        tn, fp, fn, tp = confusion_matrix(y, y_pred).ravel()

        sens = tp / (tp + fn) if (tp + fn) > 0 else 0
        spec = tn / (tn + fp) if (tn + fp) > 0 else 0

        print("Confusion matrix:")
        print(f"TN={tn} | FP={fp} | FN={fn} | TP={tp}")
        print(f"Sensibilidad: {sens:.4f}")
        print(f"Especificidad: {spec:.4f}")

    # ====================================================
    # MODELO L1 (NO DEPENDE DE K)
    # ====================================================

    for C in [0.01, 0.1, 1]:

        print(f"\n--- LR + L1 (C={C}) ---")

        pipe = Pipeline([
            ("variance", VarianceThreshold(0.01)),
            ("scaler", StandardScaler()),
            ("model", LogisticRegression(
                penalty="l1",
                solver="saga",
                C=C,
                max_iter=5000,
                class_weight="balanced"
            ))
        ])

        # ----------------------------
        # 1. CROSS-VALIDATION
        # ----------------------------
        start = time.time()

        scores = cross_validate(
            pipe,
            X_log,
            y,
            cv=cv,
            scoring={
                "roc_auc": "roc_auc",
                "f1": "f1",
                "accuracy": "accuracy"
            },
            n_jobs=2
        )

        end = time.time()

        print(
            f"AUC: {np.mean(scores['test_roc_auc']):.4f} | "
            f"F1: {np.mean(scores['test_f1']):.4f} | "
            f"ACC: {np.mean(scores['test_accuracy']):.4f} | "
            f"Tiempo: {end-start:.2f}s"
        )

        # ----------------------------
        # 2. MATRIZ DE CONFUSIÓN
        # ----------------------------
        y_pred = cross_val_predict(pipe, X_log, y, cv=cv)

        tn, fp, fn, tp = confusion_matrix(y, y_pred).ravel()

        sens = tp / (tp + fn) if (tp + fn) > 0 else 0
        spec = tn / (tn + fp) if (tn + fp) > 0 else 0

        print("Confusion matrix:")
        print(f"TN={tn} | FP={fp} | FN={fn} | TP={tp}")
        print(f"Sensibilidad: {sens:.4f}")
        print(f"Especificidad: {spec:.4f}")
```
